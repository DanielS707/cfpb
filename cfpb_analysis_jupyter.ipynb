```python
import pandas as pd
file_name = 'cfpb.xlsx'
cfpb_data = pd.read_excel(file_name, header = 8)
```

```python
# Rename columns using the first row
cfpb_data.columns = cfpb_data.iloc[0]
cfpb_data.head()
```

```python
# Drop the first row since it's now the header
cfpb_data = cfpb_data.drop(cfpb_data.index[0])

# Reset the index
cfpb_data.reset_index(drop=True, inplace=True)

# Display the first few rows with the new column names
cfpb_data.head(5)
```

```python
cfpb_data.info
```

```python
#drop the row/entry with all missing values
cfpb_data.dropna(how='all', inplace=True)
```

```python
# Check for duplicate rows
duplicate_rows = cfpb_data.duplicated().sum()
duplicate_rows
```

```python
# Check for missing values in the dataset
missing_values = cfpb_data.isnull().sum()

# Sort the missing values in descending order
sorted_missing_values = missing_values.sort_values(ascending=False)

# Display the sorted missing values
print(sorted_missing_values)
```

```python
# Drop columns where all values are missing
cfpb_data_cleaned = cfpb_data.dropna(axis=1, how='all')

# Display the shape before and after to see the changes
print("Original shape:", cfpb_data.shape)
print("Shape after dropping columns with all missing values:", cfpb_data_cleaned.shape)
```

```python
cfpb_data_head = cfpb_data_cleaned.head()

cfpb_data_cleaned.to_excel('cflb_nomissing.xlsx', index=False)
```
